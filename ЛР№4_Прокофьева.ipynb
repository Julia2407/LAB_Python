### Реализация stack и deque

```python
class Deque:
    
    def __init__(self):
        self.items = []
        

    def isEmpty(self):
        return self.items == []

    def addFront(self, item): # добавление в начало
        self.items.append(item)

    def addRear(self, item): # добавление в конец
        self.items.insert(0,item)

    def removeFront(self):
        return self.items.pop()

    def removeRear(self):
        return self.items.pop(0)
    
    def push(self, item):
         self.items.insert(0,item)
            
    def peek(self):
         return self.items[0]
        
    
    
    
```

```python
class Stack:
    
    def __init__(self):
        self.items = []
        self.size = 0
        
        
    def isEmpty(self):
        return self.items == []

    def push(self, item):
         self.items.insert(0,item)

    def pop(self):
         return self.items.pop(0)

    def peek(self):
         return self.items[0]
    
    def __len__ (self):
        return self.size
        
   
```
